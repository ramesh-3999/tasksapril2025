**Get Dummies**
*This provide the meaning full insights of the data*

```python
data = {
    'CustomerName': ['Ramesh', 'Riyansh', 'Mahesh', 'Panya', 'Ahanya', 'Juhita', 'Juwas', 'Rakshit', 'Rishi', 'Tamansai'],
    'CustomerCode': [3697, 1684, 6866, 4295, 5442, 9691, 2575, 3247, 2935, 3222],
    'Country': ['France', 'UK', 'UK', 'Canada', 'Canada', 'Germany', 'Japan', 'Japan', 'Japan', 'Germany'],
    'Business': ['IT Services', 'Manufacturing', 'Consulting', 'Education', 'Real Estate', 'IT Services', 'Consulting', 'Logistics', 'Logistics', 'Finance']
}
customer_data=pd.DataFrame(data,columns=['CustomerName','CustomerCode','Country','Business'])
print(customer_data)
```

```python
#From the above cucustomer_data we observe the country and business will come under the categorical data.
#Lets use get_dummies function --which converts the grouped column elements in to 1 and 0  and make those elements as column names
customer_data = pd.get_dummies(
    customer_data,
    columns=['Country', 'Business'],
    dtype=int,
    drop_first=True,
    prefix='',             # no prefix
    prefix_sep=''          # no separator
)
customer_data
#del customer_data
#help(pd.get_dummies)
```

```python
customer_data.drop('CustomerCode', axis=1, inplace=True)
customer_data
```

**Concatination**

```python
india_cars=pd.DataFrame({"City":['Mumbai','Hyderabad','Bengalore'],
                          "Brands":['Tata','Hyundai','Mahindra'],
                         "Speed_lmt":[90,80,80]})
india_cars                   
```

```python
Us_cars=pd.DataFrame({"City":['Newyork','Atlanta','Dallas','Dallas'],
                          "Brands":['Tesla','MG','Tata','Tata'],
                         "Speed_lmt":[120,110,130,140]}) 
Us_cars
```

```python
cars=pd.concat([india_cars,Us_cars])
cars
```

```python
#Observer that index  has repeated in above O/P as per 2 Data frames
cars=pd.concat([india_cars,Us_cars],ignore_index=True)
cars
```

```python
cars=cars.drop_duplicates()#Tis will delete all duplicate rows 
cars
```

```python
cars=cars.drop_duplicates(subset=["City"],keep="last")
cars
#observe below O/P it delted the Dallas recordwith speed 130 as its index is first than 140  speed
```
